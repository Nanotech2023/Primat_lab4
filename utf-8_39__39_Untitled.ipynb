<br> S2 – в системе 2 бревна, две машины заняты рубкой;

# Математическая модель стационарного режима
Пусть
pk – стационарная вероятность пребывания системы в состоянии Sk, k = 0,1,2,3.Тогда имеет место следующая система алгебраических уравнений, описывающая стационарный режим:
<a href="https://www.codecogs.com/eqnedit.php?latex=\begin{cases}&space;&&space;-\lambda&space;p0&space;&plus;&space;\mu&space;p1=&space;0\\&space;&&space;\lambda&space;p0&space;-&space;(\lambda&space;&plus;&space;\mu)p1&space;&plus;&space;2\mu&space;p2=&space;0&space;\\&space;&&space;\lambda&space;p1&space;-&space;(\lambda&space;&plus;&space;\mu)p2&space;&plus;&space;2\mu&space;p3=&space;0&space;\\&space;&&space;\lambda&space;p2&space;-&space;2\mu&space;p3=&space;0&space;\end{cases}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\begin{cases}&space;&&space;-\lambda&space;p0&space;&plus;&space;\mu&space;p1=&space;0\\&space;&&space;\lambda&space;p0&space;-&space;(\lambda&space;&plus;&space;\mu)p1&space;&plus;&space;2\mu&space;p2=&space;0&space;\\&space;&&space;\lambda&space;p1&space;-&space;(\lambda&space;&plus;&space;\mu)p2&space;&plus;&space;2\mu&space;p3=&space;0&space;\\&space;&&space;\lambda&space;p2&space;-&space;2\mu&space;p3=&space;0&space;\end{cases}" title="\begin{cases} & -\lambda p0 + \mu p1= 0\\ & \lambda p0 - (\lambda + \mu)p1 + 2\mu p2= 0 \\ & \lambda p1 - (\lambda + \mu)p2 + 2\mu p3= 0 \\ & \lambda p2 - 2\mu p3= 0 \end{cases}" /></a>
 Эту систему следует решать вместе с условием нормировки ___p0 + p1 + p2 + p3 = 1___. Поэтому выражаем p2 = 1 - p0 - p1 - p3 и заменяем в последнем уровнении. В итоге получаем
<a href="https://www.codecogs.com/eqnedit.php?latex=\begin{cases}&space;&&space;-\lambda&space;p0&space;&plus;&space;\mu&space;p1=&space;0\\&space;&&space;\lambda&space;p0&space;-&space;(\lambda&space;&plus;&space;\mu)p1&space;&plus;&space;2\mu&space;p2=&space;0&space;\\&space;&&space;\lambda&space;p1&space;-&space;(\lambda&space;&plus;&space;\mu)p2&space;&plus;&space;2\mu&space;p3=&space;0&space;\\&space;&&space;-\lambda&space;p0&space;-\lambda&space;p1&space;&plus;&space;(-\lambda-2\mu)p3&space;=&space;0&space;\end{cases}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\begin{cases}&space;&&space;-\lambda&space;p0&space;&plus;&space;\mu&space;p1=&space;0\\&space;&&space;\lambda&space;p0&space;-&space;(\lambda&space;&plus;&space;\mu)p1&space;&plus;&space;2\mu&space;p2=&space;0&space;\\&space;&&space;\lambda&space;p1&space;-&space;(\lambda&space;&plus;&space;\mu)p2&space;&plus;&space;2\mu&space;p3=&space;0&space;\\&space;&&space;-\lambda&space;p0&space;-\lambda&space;p1&space;&plus;&space;(-\lambda-2\mu)p3&space;=&space;0&space;\end{cases}" title="\begin{cases} & -\lambda p0 + \mu p1= 0\\ & \lambda p0 - (\lambda + \mu)p1 + 2\mu p2= 0 \\ & \lambda p1 - (\lambda + \mu)p2 + 2\mu p3= 0 \\ & -\lambda p0 -\lambda p1 + (-\lambda-2\mu)p3 = 0 \end{cases}" /></a>


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
l = 8.63
mu = 2.2
r = 2
m = 1
v = np.array([0., 0., 0.,  -l])
```

```python
matrix = np.array([
    [-l, mu, 0., 0.],
    [l, -(l+mu), 2.*mu, 0.],
    [0., l, -(l+2.*mu), 2.*mu],
    [-l, -l, 0., (-l-2.*mu)]
])
matrix
```

```python
w = np.linalg.solve(matrix, v)
w
```

Стационарные вероятности в данном случае равны p1 = 0.03609177, p1 = 0.14157817, p2 = 0.27768628, p3 = 0.54464377.

# Стационарные характеристики СМО

На основе полученных значений вероятностей пребывания системы в состояниях
определяются требуемые показатели эффективности стационарного режима. А именно,
<li>___p0 = 0.03609177___ – вероятность того, что все машины не работают;
<li>___p1 = 0.14157817___ – вероятность того, что одна машина занята рубкой;
<li>___p2 = 0.27768628___ – вероятность того, что две машины заняты рубкой;
<li>___p3 = 0.54464377___ – вероятность того, что две машины занята рубкой и одно бревно ожидает в очереди;

```python
X0 = w[0] # машины свободны
X1 = w[1] # 1 машина занята
X2 = w[2] + w[3] # 2 машины заняты

M_x = 0*X0 + 1*X1 + 2*X2
print('Cреднее число работающих машин равно ' + str(M_x))

Y0 = w[2] + w[3] # 2 машины заняты
Y1 = w[1] # 1 машина занята
Y2 = w[0] # машины свободны

M_y = 0*Y0 + 1*Y1 + 2*Y2
print('Cреднее число простаивающих машин равно ' + str(M_y))

k = M_x/r
print('Коэффициент загрузки машин равен ' + str(k))

k = M_y/r
print('Коэффициент простоя машин равен ' + str(k))

Z0 = w[0] + w[1] + w[2] # 0 бревен в очереди
Z1 = w[3] # 1 бревно в очереди

M_z = 0*Z0 + 1*Z1 
print('Cреднее число бревен, находящихся в очереди на рубку, равно ' + str(M_z))
```

# Математическая модель нестационарного режима

Обозначим через p(t)k вероятность пребывания системы в момент времени t в состоянии Sk, k = 0,1,2,3. 
Нестационарный режим описывается следующей системой дифференциальных уравнений
<a href="https://www.codecogs.com/eqnedit.php?latex=\begin{cases}&space;&&space;p0{'}(t)&space;=-\lambda&space;p0(t)&space;&plus;&space;\mu&space;p1(t)\\&space;&&space;p1{'}(t)&space;=\lambda&space;p0(t)&space;-&space;(\lambda&space;&plus;&space;\mu)p1(t)&space;&plus;&space;2\mu&space;p2(t)&space;\\&space;&&space;p2{'}(t)&space;=\lambda&space;p1(t)&space;-&space;(\lambda&space;&plus;&space;2\mu)p2(t)&space;&plus;&space;2\mu&space;p3(t)&space;\\&space;&&space;p3{'}(t)&space;=\lambda&space;p2(t)&space;-&space;2\mu&space;p3(t)&space;\end{cases}" target="_blank"><img src="https://latex.codecogs.com/gif.latex?\begin{cases}&space;&&space;p0{'}(t)&space;=-\lambda&space;p0(t)&space;&plus;&space;\mu&space;p1(t)\\&space;&&space;p1{'}(t)&space;=\lambda&space;p0(t)&space;-&space;(\lambda&space;&plus;&space;\mu)p1(t)&space;&plus;&space;2\mu&space;p2(t)&space;\\&space;&&space;p2{'}(t)&space;=\lambda&space;p1(t)&space;-&space;(\lambda&space;&plus;&space;2\mu)p2(t)&space;&plus;&space;2\mu&space;p3(t)&space;\\&space;&&space;p3{'}(t)&space;=\lambda&space;p2(t)&space;-&space;2\mu&space;p3(t)&space;\end{cases}" title="\begin{cases} & p0{'}(t) =-\lambda p0(t) + \mu p1(t)\\ & p1{'}(t) =\lambda p0(t) - (\lambda + \mu)p1(t) + 2\mu p2(t) \\ & p2{'}(t) =\lambda p1(t) - (\lambda + 2\mu)p2(t) + 2\mu p3(t) \\ & p3{'}(t) =\lambda p2(t) - 2\mu p3(t) \end{cases}" /></a>
Решение этой системы должно удовлетворять начальному условию p0(0) = 1, p1(0) = 0, p2(0) = 0, p3(0) = 0

```python
from scipy.integrate import solve_ivp
time = (0, 3)
start = [1, 0, 0, 0]
def rhs(s, p):
    return [
            -l*p[0]+mu*p[1], 
            l*p[0]-(mu+l)*p[1]+2*mu*p[2], 
            l*p[1]-(2*mu+l)*p[2]+2*mu*p[3],
            l*p[2]-2*mu*p[3]
           ]
res = solve_ivp(rhs, time, start)
```

```python
plt.figure(figsize=(12,5), dpi=100)
plt.plot(res.t, res.y.T)
plt.title('Вероятности состояний системы')
```

```python
print('Нестационарные знвачения - ' + str(res.y.T[-1]))
print('Стационарные знвачения - ' + str(w))
```

Вероятности p(t)k с ростом времени t стремятся к своим стационарным значениям pk, независящим от времени

# Нестационарные характеристики СМО

```python
k1 = np.array([])
k2 = np.array([])

for i in res.y.T:
    k1 = np.append(k1,(i[1] + 2*i[0])/2)
    k2 = np.append(k2,(i[1] + 2*(i[2] + i[3]))/2)
```

```python
plt.figure(figsize=(12,5), dpi=100)
plt.plot(res.t, k1, label='коэф. простоя')
plt.plot(res.t, k2, label='коэф. загрузки')
plt.title('Коэффициенты загрузки и простоя машин в цехе')
plt.legend()
```

Видим, что функции k_x(t) и k_y(t) с течением времени приближаются к своим стационарным значениям, равным 0.893 и 0,106 соответственно
